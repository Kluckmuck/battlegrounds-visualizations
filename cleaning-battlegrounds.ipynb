```python
df = pd.read_csv(folder + "BGMatchDataBoards.csv")
df.dtypes
```

```python
## Creating separate features for each minion
df2 = df['minions'].str.split(r'[)][,]',expand=True)
df2 = df2.rename(columns={0:"minion1", 1:"minion2", 2:"minion3", 3:"minion4", 4:"minion5", 5:"minion6", 6:"minion7"})
df = df.merge(df2, left_index=True,right_index=True)
df = df.drop(['minions'], axis=1)

df
```
